```python
import pandas as pd
import numpy as np
```

```python
listings_ams = pd.read_csv('../data/cleaned/listings_ams2.csv', index_col=0)
```

```python
unique_amenities = {}

for entry in listings_ams["amenities"]:    
    s = entry[1:-1]
    l = [x.replace("\"", "").strip() for x in s.split(",")]
    
    for amenity in entry:
        if amenity not in unique_amenities:
            unique_amenities[amenity] = 1
        else:
            unique_amenities[amenity] += 1

amenities = unique_amenities.items()
```

```python
def place_has_or_not(l,string):
    if any(string in l for x in l):
        return 1
    else:
        return 0
```
