#CONSIDERATIONS:
The group room_type has a problem of class imbalance. 
In order to improve the model, one way is to consider only the room_type important for the problem
(entire property or private room).
Considering also 2 numerical variables: bedrooms and accommodates

KNN - #neighbors, weights, p

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error 
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score
from sklearn.preprocessing import StandardScaler


```

```python
listings_ams_all=pd.read_csv('../data/cleaned/listings_ams_all.csv', index_col=0)
```

```python
categorical = pd.read_csv('../data/cleaned/categorical.csv', index_col=0)
```

```python
categorical = categorical[categorical['room_type']==('Entire home/apt' or 'Private room')]
categorical.shape
```

```python
listings_ams_all= listings_ams_all[listings_ams_all['room_type']==('Entire home/apt' or 'Private room')]
listings_ams_all.shape
```

```python
listings_ams_all.shape
```

```python
num = listings_ams_all.loc[:,['bedrooms', 'accommodates']]
```

```python
num.shape
```

```python
categorical = pd.merge(categorical, num, left_index=True, right_index=True)
```

```python
categorical = categorical.drop(['host_identity_verified'], axis=1)
```

```python
categorical.shape
```

```python
#X-y split
X = categorical
y = listings_ams_all['price']
```

```python
X_n = categorical.loc[:,['bedrooms', 'accommodates']]
X_c = categorical.drop(['bedrooms', 'accommodates'], axis=1)
```

```python
X_c
```

```python
X_n = X_n.reset_index(drop=True)
```

```python
encoder = OneHotEncoder(drop='first')
encoder.fit(X_c)

X_c_np = encoder.transform(X_c).toarray()

X_c = pd.DataFrame(X_c_np, columns=encoder.get_feature_names_out())

X_c.shape
X_c
```

```python
transformer = StandardScaler() 
transformer.fit(X_n) 

X_n_np = transformer.transform(X_n)
X_n_scaled = pd.DataFrame(X_n_np, columns=X_n.columns, index=X_n.index)

X_n_scaled.head()
```

```python
X=pd.merge(X_n, X_c, left_index=True, right_index=True) # concat numerical and categorical transformations
X.shape
#X.tail()
```

```python
#train-test-split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=11)
```

```python

knn_models = {} # keys = k_distance_p, value = model
weights = ["uniform","distance"]
p_values = [1,2,3]
scores = {}
for k in range(2,15):
    for weight in weights:
        for value in p_values:
            model = KNeighborsRegressor(n_neighbors=k,weights=weight,p=value)
            print("Training the k-NN model with {} neighbours with weights {} and p set to {}...".format(k,weight,value))
            model.fit(X_train, y_train)
            #knn_models.append(model)
            knn_models[str(k)+"_"+weight+"_"+str(value)] = model
            scores[str(k)+"_"+weight+"_"+str(value)] = model.score(X_test, y_test)
            print()
```

```python
knn_models.keys()
```

```python
knn_scores = {}

for key in knn_models.keys():
    knn_scores[key] = knn_models[key].score(X_test, y_test)

knn_scores
```

```python
for key in list(scores.keys()):
    print("The R2 of the k-nn model with {} neighbours, weights set to {}, and p set to {} on the TEST set was: {:.3f}".format(key.split("_")[0],key.split("_")[1],key.split("_")[2],scores[key]))
```
