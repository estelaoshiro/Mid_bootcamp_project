#CONSIDERATIONS:
Linear Regression Model - only categorical features

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error 
from sklearn.neighbors import KNeighborsRegressor
```

```python
listings_ams_all=pd.read_csv('../data/cleaned/listings_ams_all.csv', index_col=0)
```

```python
listings_ams_all.isna().sum()
```

```python
categorical = listings_ams_all.select_dtypes(object)
```

```python
categorical
```

```python
def clean_cat(x):
    if x == 't':
        return 1
    elif x == 'f':
        return 0
```

```python
categorical['host_is_superhost']=categorical['host_is_superhost'].apply(clean_cat)
```

```python
categorical['host_identity_verified']=categorical['host_identity_verified'].apply(clean_cat)
```

```python
categorical['has_availability']=categorical['has_availability'].apply(clean_cat)
```

```python
categorical['instant_bookable']=categorical['instant_bookable'].apply(clean_cat)
```

```python
categorical = pd.merge(categorical, listings_ams_all.loc[:,['parking', 'cooking_basics',
       'dishwasher', 'tv', 'iron', 'private_patio_or_balcony', 'freezer',
       'dryer', 'private_entrance', 'microwave', 'self_check_in',
       'pets_allowed', 'bbq_grill']], left_index=True, right_index=True)


display(categorical.shape)
categorical.head()
```

```python
categorical=categorical.drop(['name', 'amenities','bathrooms_text','has_availability', 'property_type'], axis=1)
```

```python
categorical['neighbourhood_cleansed'].unique()
```

```python
def group_neighbourhood(x):
    if x in ['Bijlmer-Oost','Bijlmer-Centrum','Gaasperdam - Driemond', 'De Aker - Nieuw Sloten', 'Osdorp',
            'Buitenveldert - Zuidas','Geuzenveld - Slotermeer']:
        return 'Other'
    else:
        return x
```

```python
categorical['neighbourhood_cleansed']=categorical['neighbourhood_cleansed'].apply(group_neighbourhood)
```

```python
categorical['neighbourhood_cleansed'].nunique()
```

```python
categorical
```

```python
categorical.to_csv('../data/cleaned/categorical.csv')
```

```python
#X-y split
X = categorical
y = listings_ams_all['price']
```

```python
#train-test-split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=11)
```

```python
encoder = OneHotEncoder(drop='first')
encoder.fit(X_train)

X_train_cat_np = encoder.transform(X_train).toarray()
X_test_cat_np  = encoder.transform(X_test).toarray()

X_train_cat = pd.DataFrame(X_train_cat_np, columns=encoder.get_feature_names_out())
X_test_cat  = pd.DataFrame(X_test_cat_np,  columns=encoder.get_feature_names_out())

X_train_cat.head()
```

```python
#MODEL 1 - LINEAR REGRESSION
lr=LinearRegression()    
lr.fit(X_train_cat, y_train)  
y_pred_lr=lr.predict(X_test_cat)
```

```python
#EVALUATION
#R2
print ('train R2: {} vs test R2: {}'.format(lr.score(X_train_cat, y_train),
                                            lr.score(X_test_cat, y_test)))
```

```python
#MSE
mse=mean_squared_error
train_mse=mse(lr.predict(X_train_cat), y_train)
test_mse=mse(lr.predict(X_test_cat), y_test)

print ('train MSE: {} vs test MSE: {}'.format(train_mse, test_mse))
```

```python
#RMSE
print ('train RMSE: {} vs test RMSE: {}'.format(train_mse**.5, test_mse**.5))
```

```python
#MAE
mae=mean_absolute_error
train_mae=mae(lr.predict(X_train_cat), y_train)
test_mae=mae(lr.predict(X_test_cat), y_test)

print ('train MAE: {} vs test MAE: {}'.format(train_mae, test_mae))
```
