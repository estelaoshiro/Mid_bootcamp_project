#CONSIDERATIONS:
The group room_type has a problem of class imbalance. 
In order to improve the model, one way is to consider only the room_type important for the problem
(entire property and private rooms).
Considering also 2 numerical variables: bedrooms and accommodates
Linear Model

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error 

```

```python
listings_ams_all=pd.read_csv('../data/cleaned/listings_ams_all.csv', index_col=0)
```

```python
categorical = pd.read_csv('../data/cleaned/categorical.csv', index_col=0)
```

```python
categorical = categorical[categorical['room_type']==('Entire home/apt' or 'Private room')]
categorical.shape
```

```python
listings_ams_all= listings_ams_all[listings_ams_all['room_type']==('Entire home/apt' or 'Private room')]
listings_ams_all.shape
```

```python
listings_ams_all.shape
```

```python
num = listings_ams_all.loc[:,['bedrooms', 'accommodates']]
```

```python
num.shape
```

```python
categorical = pd.merge(categorical, num, left_index=True, right_index=True)
```

```python
categorical = categorical.drop(['host_identity_verified'], axis=1)
```

```python
categorical.shape
```

```python
#X-y split
X = categorical
y = listings_ams_all['price']
```

```python
X_n = categorical.loc[:,['bedrooms', 'accommodates']]
X_c = categorical.drop(['bedrooms', 'accommodates'], axis=1)
```

```python
X_c
```

```python
X_n = X_n.reset_index(drop=True)
```

```python
encoder = OneHotEncoder(drop='first')
encoder.fit(X_c)

X_c_np = encoder.transform(X_c).toarray()

X_c = pd.DataFrame(X_c_np, columns=encoder.get_feature_names_out())

X_c.shape
X_c
```

```python
X=pd.merge(X_n, X_c, left_index=True, right_index=True) # concat numerical and categorical transformations
X.shape
#X.tail()
```

```python
#train-test-split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=11)
```

```python
#MODEL 1 - LINEAR REGRESSION
lr=LinearRegression()    
lr.fit(X_train, y_train)  
y_pred_lr=lr.predict(X_test)
```

```python
X_train
```

```python
#EVALUATION
#R2
print ('train R2: {} vs test R2: {}'.format(lr.score(X_train, y_train),
                                            lr.score(X_test, y_test)))
```

```python
#MSE
mse=mean_squared_error
train_mse=mse(lr.predict(X_train), y_train)
test_mse=mse(lr.predict(X_test), y_test)

print ('train MSE: {} vs test MSE: {}'.format(train_mse, test_mse))
```

```python
#RMSE
print ('train RMSE: {} vs test RMSE: {}'.format(train_mse**.5, test_mse**.5))
```

```python
#MAE
mae=mean_absolute_error
train_mae=mae(lr.predict(X_train), y_train)
test_mae=mae(lr.predict(X_test), y_test)

print ('train MAE: {} vs test MAE: {}'.format(train_mae, test_mae))
```
